```python
# Import the necessary packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
cern = pd.read_csv("C:\\Users\\Hussen\\Desktop\\Data\\dielectron.csv") # Import the data
cern.head()
```

```python
# Give x and y the data they need
x = cern['pz1']
y = cern['eta1']

plt.figure(figsize=(8,5))  # Adjust the size of the figure
plt.scatter(x,y)  # Plot the scatterplot

#  Plot the different models of our fit
import math
x = np.linspace(-500,500,400)
y1 = np.array([1.7565*math.tanh(0.0913*i) for i in x])
y2 = (0.24752*x)/(2.109857+(0.13515*x)**2)**0.5
y3 = 0.686672*x/(2.505943+abs(0.327373*x))
plt.plot(x,y1,label='y = 1.75tanh(0.09x)',color='r')
plt.plot(x,y2,label='y = 0.247x/sqrt(2.1+(0.13X)^2)',color='g')
plt.plot(x,y3,label='y = 0.686x/(2.5+abs(0.327x))',color='y')
plt.xlabel('pz1', labelpad=15, fontweight='bold', fontsize=13.5)
plt.ylabel('eta1', labelpad=15, fontweight='bold', fontsize=13.5)
plt.legend()
# plt.savefig('cern2.jpg',dpi=300)

plt.show()
```

```python
# Give x and y the data they need
x = cern['pz1']
y = cern['eta1']

plt.figure(figsize=(8,5))  # Adjust the size of the figure
plt.scatter(x,y)  # Plot the scatterplot

# Plot the models without being adjusted
x = np.linspace(-500,500,400)
y1 = np.array([math.tanh(i) for i in x])
y2 = (x)/(1+(x)**2)**0.5
y3 = x/(1+abs(x))
plt.plot(x,y1,label='y = tanh(x)',color='r')
plt.plot(x,y2,label='y = x/sqrt(1+(X)^2)',color='g')
plt.plot(x,y3,label='y = x/(1+abs(x))',color='y')
plt.xlabel('pz1', labelpad=15, fontweight='bold', fontsize=13.5)
plt.ylabel('eta1', labelpad=15, fontweight='bold', fontsize=13.5)
plt.legend()
# plt.savefig('cern3.jpg',dpi=300)

plt.show()
```
