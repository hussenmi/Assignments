```python
# The python code here will calculate the best distance and the best order of cities to go through to optimize our 
# objective function (distance), but it should be noted that the program creates these cities 'randomly', so the distance 
# distance between them might not be the same as in the real world. There are also other parameters chosen such as the 
# mutation rate, survivor rate, and number of generations it has to go through

# Import the modules we use in this notebook

import matplotlib.pyplot as plt
import random
import numpy as np
import networkx as nx


# This generates a map of the cities, meaning that each city will be given x and y coordinates. The arguments, xmin, xmax, 
# ymin, and ymax specify what the maximin and minimum position in the x and y axes the cities could have, and the 'seed' 
# insures thatwe get the same result no matter who runs the function. So even though the position if the cities is randomly 
# generated, when we share the notebook with other people, they also get the same result.

def generate_uniform_map(xmin = -100, xmax = 100, 
                     ymin = -100, ymax =100, 
                     cities_count = 10, plot = False, seed = 2021):
    np.random.seed(seed)
    xs = np.random.uniform(xmin, xmax, cities_count)
    ys = np.random.uniform(ymin, ymax, cities_count)

    if plot:
        plt.scatter(xs, ys, s = 200, c = 'darkmagenta')
        plt.xlabel('X coordinate of the city')
        plt.ylabel('Y coordinate of the city')
        plt.show()

    # Pairs each x and y
    coords = np.dstack((xs, ys))[0]
    cities = dict()

    for i in range(len(coords)):
        cities[str(i)] = coords[i]

    return cities

# From the list of cities, we create a population filled with random routes, and we use a random shuffle method because we 
# don't want to get the same number more than once, which would mean that we are re-visiting a city. The randomness also
# helps us get a diverse population, which will prevent us from getting an early convergence

def random_route(citylist):
    keys = list(citylist.keys())
    np.random.shuffle(keys)
    return keys

def initialize_routes(population_size, citylist):
    population = []
    for i in range(population_size):
        population.append(random_route(citylist))
    return population



# We first get a distance between two cities using the 'get_distance' function and we get the distance of the whole route 
# using the 'get_route_length' function

def get_distance(citylist, indexi, indexj):
    city_a = citylist[indexi]
    city_b = citylist[indexj]
    x_dis = abs(city_a[0] - city_b[0])
    y_dis = abs(city_a[1] - city_b[1])
    distance = np.sqrt((x_dis ** 2) + (y_dis ** 2))  # Use pythagorean theorm to find the distance
    return distance

def get_route_length(citylist, route):
    total_dist = 0
    for index in range(len(route)):
        if index < len(route) - 1:
            dist = get_distance(citylist, route[index], route[index+1])
        else:
            dist = get_distance(citylist, route[index], route[0])
        total_dist += dist
    return total_dist


# Here, we rank our population based on the distance we just calculated. The distance, therefore, serves as the fitness for
# this problem. The ones with the shortest distance will be ranked higher

def rank_routes(citylist, population):
    fitnessresults = []

    for index in range(len(population)):
        fitnessresults.append([index, get_route_length(citylist, population[index])])

    fitnessresults.sort(key=lambda x: x[1])
    return fitnessresults  


# Here, we select our parents that we will apply crossover on. This depends on our fitness results, population size, and 
# survival rate

def get_crossover_pool(fitnessresults, population_size, survival_rate):
    cross_pool = []
    elite_amount = int(population_size * survival_rate)
    
    # We first place our direct survivors into the pool
    for route in fitnessresults[0:elite_amount]:
        cross_pool.append(route[0])

    return cross_pool


# Here, we apply the Davis' Crossover because that is the one that is most fit when dealing with problems that involve 
# permutations. It works by taking a portion from one parent's chromosome and from the other parent, it takes the ones that 
# aren't already included in the offspring from the first parent and it takes themin order. This is really important because
# we don't want to repeat a city that we already have, and this crossover alogorithm takes care of that very well.
# The 'davis_crossover' algorithm is for creating offsprings from two parents and the 'population_crossover' is for creating
# offsprings for the population.

def davis_crossover(parent1, parent2):
    num_cities = len(parent1)
    child_1 = []
    child_2 = []
    breakpoints = np.sort(np.random.choice(num_cities,2,replace=False))
    for i in range(breakpoints[0],breakpoints[1]):
        child_1.append(parent1[i])
        child_2.append(parent2[i])
    sortpoint = num_cities - breakpoints[0]
    j = breakpoints[1]
    while len(child_1) < num_cities:
        while parent2[j%num_cities] in child_1:
            j += 1
        child_1.append(parent2[j%num_cities])
    child_1 = child_1[sortpoint:] + child_1[:sortpoint]
    j = breakpoints[1]
    while len(child_2) < num_cities:
        while parent1[j%num_cities] in child_2:
            j += 1
        child_2.append(parent1[j%num_cities])
    child_2 = child_2[sortpoint:] + child_2[:sortpoint]
    return [child_1, child_2]


def population_crossover(population_size, population, cross_pool):
    children = []

    if len(cross_pool) % 2 != 0:
        cross_pool.pop()

    for routeindex in cross_pool:
        children.append(population[routeindex])

    while len(children) < population_size:
        sampleindex = np.random.choice(cross_pool, 2, replace = False)
        parent1 = population[sampleindex[0]]
        parent2 = population[sampleindex[1]] 
        children += davis_crossover(parent1, parent2)
    return children


# Here, we apply mutation to individual chromosomes or solutions by swaping random genes with others in the same chromosome
# or route. This helps us get over things such as a local extreme and helps alleviate some of the problems that might arise 
# from the lack of diversity in the population. The 'mutate' function creates a mutation in a single chromosome, and the 
# 'mutate_population' creates a mutation in the entire population.

def mutate(route, mutation_rate, seed = 2021):
    for swapped in range(len(route)):
        if(np.random.random() < mutation_rate):
            swapwith = int(np.random.random() * len(route))
            route[swapped], route[swapwith] = route[swapwith], route[swapped]
            
    return route

def mutate_population(mutation_rate, children):
    mutated_routes = []

    for route in children:
        mutated_route = mutate(route, mutation_rate)
        mutated_routes.append(mutated_route)

    return mutated_routes



# These are just visualization functions. They don't affect any of the code pertaining to the genetic algorithm 

def visualize(citylist, population, fitnessresults):
    # Creating a graph
    graph = nx.Graph()
    # Adding nodes to the graph
    graph.add_nodes_from(citylist.keys())
    # Getting the current best route
    if fitnessresults:
        best_route = fitnessresults[0][0]
        # Adding the edges that correspond to the current best route
        for node_index in range(len(population[best_route]) - 1):
            graph.add_edge(str(population[best_route][node_index]), str(population[best_route][node_index+1]))
        graph.add_edge(str(population[best_route][-1]), str(population[best_route][0]))
    # Creating layout
    pos = {}
    plt.figure(figsize=(6.5,6.5))
    for node in citylist.keys():
        pos[node] = tuple(citylist[node])
    nx.drawing.nx_pylab.draw_networkx(graph,pos=pos,font_color='white',\
        font_weight='bold',font_size=11,node_color='darkmagenta',node_size=300)
    plt.show()

# The function below chooses what generations to display
# The routes change quickly at first, and then more slowly as the evolution progresses.
# Thus, this function chooses to look at generations that are closer together 
# earlier in the evolution, becoming more spaced out as we go along. 
def get_displays(n_displays, n_plots, n_generations):
    displayed_gens = [1+np.floor((n_generations-1)*(i/n_displays)**2) for i in range(0, n_displays+1)]
    displayed_plots = [1+np.floor((n_generations-1)*(i/n_plots)**2) for i in range(0, n_plots+1)]
    return displayed_gens, displayed_plots


# Define the parameters of our genetic algorithm

cities_count = 15  # We have 15 cities we need to visit
population_size = 40  # Let's populate our number of solutions as 40
n_generations = 40  # Let's have 40 generations
mutation_rate = 0.02
survival_rate = 0.6


# Initialize the population by calling the functions with the right arguments
cities = generate_uniform_map(cities_count = cities_count, seed = 2021)
population = initialize_routes(population_size, cities)
fitnessresults = rank_routes(cities, population)

# Get generations to display
default_displays = 3
display = get_displays(default_displays,default_displays,n_generations)
fitness_evolution = np.zeros(n_generations)

# Main loop of the algorithm
for gen in range(1, n_generations+1):
    if gen in display[0]:
        print(f'Generation: {gen}\nBest distance: {str(round(fitnessresults[0][1],3))}')
    if gen in display[1]:
        visualize(cities, population, fitnessresults)
    cross_pool = get_crossover_pool(fitnessresults, population_size, survival_rate)
    if len(cross_pool)<2:
        print("Not enough parents to produce children!")
        print("For a population of size",population_size,", the survival rate must be at least",2/population_size)
        break
    else:
        children = population_crossover(population_size, population, cross_pool)
        population = mutate_population(mutation_rate, children)
        fitnessresults = rank_routes(cities, population)
        fitness_evolution[gen-1] = fitnessresults[0][1]

# plot to show fitness over the generations
if len(cross_pool)>=2:
    plt.plot(list(range(n_generations)),fitness_evolution,c = 'seagreen',linewidth=2)
    plt.title('Best route length per generation', fontsize = 14)
    plt.xlabel('Generation number', fontsize = 12)
    plt.ylabel('Best route length', fontsize = 12)
    plt.show()
    print("Best path in the final generation:\n",population[0],"\nDistance:",round(fitnessresults[0][1],3))
```

```python
import scipy
import numpy as np
import matplotlib.pyplot as plt

# This model includes four categories. The additional one is the 'exposed' category. This category isn't infectious yet, but
# they have already coontracted the virus, so they areoutof the susceptible population, but not infective yet. Here, the
# rate at which exosed people become infecive is 'c', and they leave that category at a rate of c*E.
def SEIR_model(y,t,b,k,c):
    S,E,I,R = y
    
    dsdt = -b*S*I
    dedt = b*S*I - c*E
    didt = c*E - k*I
    drdt = k*I
    
    return([dsdt,dedt,didt,drdt])

# Initial conditions
So,Eo,Io,Ro = 900,20,10,0
b,k,c = 0.085, 0.062, 0.02

# Time vector
t = np.linspace(0,365,1000)


# Use the 'odeint' method from the scipy module to solve the differential equations based on initial conditions & parameters
# The inputs of the 'odeint' method are the functions and their arguments -- initial conitions, time vector, and parameters
sol = scipy.integrate.odeint(SEIR_model, [So,Eo,Io,Ro], t, args=(b,k,c))
sol = np.array(sol)

# Plot the different populations as a function of time
import matplotlib.pyplot as plt

plt.plot(t, sol[:,0],label='S(t)')
plt.plot(t, sol[:,1],label='E(t)')
plt.plot(t, sol[:,2],label='I(t)')
plt.plot(t, sol[:,3],label='R(t)')
plt.title('SEIR Model', fontweight='bold')
plt.xlabel('Time', labelpad=10, fontweight='bold', fontsize=11.5)
plt.ylabel('S,E,I,R', labelpad=10, fontweight='bold', fontsize=11.5)
plt.legend()
plt.show()
```

```python
from scipy import integrate

# This model includes four categories. The additional one is 'S2', whcih differes from S1 in that they don't have thw same
# infection rate; they don't enter the infectives category at the same rate (b1 for S1 and b2 for S2). The equations are
# then adjusted to include this changes.
def S1S2IR_model(y,t,b1,b2,k):
    S1,S2,I,R = y
    
    ds1dt = -b1*S1*I
    ds2dt = -b2*S2*I
    didt = b1*S1*I + b2*S2*I - k*I
    drdt = k*I
    
    return([ds1dt,ds2dt,didt,drdt])

# Initial conditions
S1o,S2o,Io,Ro = 900,30,10,0
b1,b2,k = 0.085, 0.001, 0.062

# Time vector
t = np.linspace(0,100,100)

# Use the 'odeint' method from the scipy module to solve the differential equations based on initial conditions & parameters
# The inputs of the 'odeint' method are the functions and their arguments -- initial conitions, time vector, and parameters
sol = scipy.integrate.odeint(S1S2IR_model, [S1o,S2o,Io,Ro], t, args=(b1,b2,k))
sol = np.array(sol)

# Plot the different populations as a function of time
import matplotlib.pyplot as plt

plt.plot(t, sol[:,0],label='S1(t)')
plt.plot(t, sol[:,1],label='S2(t)')
plt.plot(t, sol[:,2],label='I(t)')
plt.plot(t, sol[:,3],label='R(t)')
plt.title('S1S2IR Model', fontweight='bold')
plt.xlabel('Time', labelpad=10, fontweight='bold', fontsize=11.5)
plt.ylabel('S1,S2,I,R', labelpad=10, fontweight='bold', fontsize=11.5)
plt.legend()
plt.show()
```

```python
from scipy import integrate

# This model assumes the a vaccine is distributed in the population. If that happens, the change is that people can move
# from the susceptible category to the removed category without going through the infectives category. To account for that,
# we assume that the vaccination rate is constant and introduce that rate to the change of number of suscptibles and the
# number of removed categories. Here, it's important to note that, the variable 'vaccinated' represents the rate at which
# people are vaccinated. for example, here, we are assumning that 30 people are being vaccinated per unit time.
def SIR_vaccine_model(y,t,b,k,vaccinated):
    S,I,R = y
    
    dsdt = -b*S*I - vaccinated
    didt = b*S*I - k*I
    drdt = k*I + vaccinated
    
    return([dsdt,didt,drdt])

# Initial conditions
vaccinated = 30
So,Io,Ro = 900,1,0
b,k = 0.085, 0.062

# Time vector
t = np.linspace(0,20,100)

# Use the 'odeint' method from the scipy module to solve the differential equations based on initial conditions & parameters
# The inputs of the 'odeint' method are the functions and their arguments -- initial conitions, time vector, and parameters
sol = scipy.integrate.odeint(SIR_vaccine_model, [So,Io,Ro], t, args=(b,k,vaccinated))

# Filter the solution so that it doesn't keep vaccinating people even if they are not susceptible anymore, which would
# increase the number of 'removed' population above the total population size
sol = sol[sol[:,2]<So]  
sol = np.array(sol)

t = np.linspace(0,20,len(sol[:,0]))

# Plot the different populations as a function of time
import matplotlib.pyplot as plt

plt.plot(t, sol[:,0],label='S(t)')
plt.plot(t, sol[:,1],label='I(t)')
plt.plot(t, sol[:,2],label='R(t)')
plt.title('SIR Vaccine Model', fontweight='bold')
plt.xlabel('Time', labelpad=10, fontweight='bold', fontsize=11.5)
plt.ylabel('S,I,R', labelpad=10, fontweight='bold', fontsize=11.5)
plt.legend()
plt.show()
```

```python
from scipy import integrate

# This model assumes natural births and deaths occur in every category of the population, and it also doesn't assume a
# constant population; the birth and death rates are different. 'N' is the total number of population at the beginning, and 
# every new born is added to the susceptible category with a rate of 'br' and the Natural death rate for every category is 
# assumed to be dr, so we subtract a rate that is proprtional to the number of people in that category with the
# proportionality constant 'dr'.
def SIR_bd_model(y,t,b,k,br,dr):
    S,I,R = y
    N = S+I+R
    
    dsdt = -b*S*I + br*N - dr*S
    didt = b*S*I - k*I -dr*I
    drdt = k*I - dr*R
    
    return([dsdt,didt,drdt])

# Initial conditions
So,Io,Ro = 900,10,0
b,k,br,dr = 0.085, 0.062, 0.01199, 0.008

# Time vector
t = np.linspace(0,40,100)

# Use the 'odeint' method from the scipy module to solve the differential equations based on initial conditions & parameters
# The inputs of the 'odeint' method are the functions and their arguments -- initial conitions, time vector, and parameters
sol = scipy.integrate.odeint(SIR_bd_model, [So,Io,Ro], t, args=(b,k,br,dr))
sol = np.array(sol)

# Plot the different populations as a function of time
import matplotlib.pyplot as plt

plt.plot(t, sol[:,0],label='S(t)')
plt.plot(t, sol[:,1],label='I(t)')
plt.plot(t, sol[:,2],label='R(t)')
plt.title('SIR with Birth and Death Model', fontweight='bold')
plt.xlabel('Time', labelpad=10, fontweight='bold', fontsize=11.5)
plt.ylabel('S,I,R', labelpad=10, fontweight='bold', fontsize=11.5)
plt.legend()
plt.show()
```

Solving the SIR differential equations using Euler's method means that we start from the differntial equations and initial conditions, which are the initial number of susceptible, infectious, and removed people in the population and find a function thta would express these variables as a function of time. The algorithm starts with a formula for a slope (a derivative) and an initial condition (for example, y(0)), in our case, we have S(0), I(0), and R(0) as our initial conditions. From our derivative formula, we can get the value for the slope at our initial condition. We then assume the slope remains the same until our next x-value and use that to approximate our y-value. Once we get the x and y-values, we can then use the formula for the derivative to find the slope at that point. We then repeat the above process -- use our x and slope to get y and then use our x and slope to estmate the next y. And we iterate this until we reach the value of the function at our desired point. It is important to note that when we proceed to the next values, we increment our x value by a certain number (h), and the smaller the step size, the better our approximation is going to be. That is because, the Euler method works by assuming that the slope doesn't change until the next x-value, and if we choose a step size that is very small, that assumption becomes more and more accurate, which leads to more accurate results. The system itselfwe are modelling is discrete becuase the number of people is discrete, and it can only take on values of whole numbers, but the SIR model is continuous because we are treating these as variables of time and time is conitinuous, so we can put any value for time and we could get an output.
